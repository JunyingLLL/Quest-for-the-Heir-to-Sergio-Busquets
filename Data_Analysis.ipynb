```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Busquets vs Barcelona Players

```python
original_df = pd.read_html('https://fbref.com/en/squads/206d90db/2022-2023/all_comps/Barcelona-Stats-All-Competitions')
```

```python
barca_df = original_df[0]
```

```python
barca_df.columns = barca_df.columns.droplevel(0)
barca_df.drop(barca_df.tail(22).index, inplace = True)
barca_df.drop(barca_df.index[0], inplace = True)
```

```python
barca_df.reset_index(inplace = True)
barca_df.loc[7, "Pos"] = "MF"
barca_df.loc[13, "Pos"] = "FW"
```

```python
barca_df
```

```python
barca_df.columns
```

```python
barca_df["PrgPassesper90"] = barca_df["PrgP"] / barca_df["90s"]
```

```python
barca_df = barca_df.sort_values(by = ["PrgPassesper90"], ascending = False)

sns.set(style="darkgrid")
prog_passes_plot = sns.barplot(data = barca_df, x = "PrgPassesper90", y = "Player", hue = "Pos")
prog_passes_plot.set(ylabel = None)
plt.xlabel("Progressive Passes per 90")
plt.title("Barcelona Players Progressive Passes in 2022-23", pad = 15, color = "#333333", weight = "bold")
plt.show()
```

```python
barca_defend = original_df[43]
barca_defend.columns = barca_defend.columns.droplevel(0)
barca_defend.drop(barca_defend.tail(12).index, inplace = True)
barca_defend.drop(barca_defend.index[0], inplace = True)
barca_defend.reset_index(inplace = True)
barca_defend.loc[7, "Pos"] = "MF"
barca_defend.loc[13, "Pos"] = "FW"
```

```python
barca_defend
```

```python
barca_defend.columns
```

```python
barca_defend.columns.values[11] = "DribTkl"
barca_defend.columns
```

```python
barca_defend = barca_defend.sort_values(by = ["Tkl+Int"], ascending = False)
# barca_defend = barca_defend.sort_values(by = ["Tkl%"], ascending = False)

sns.set(style="darkgrid")
defensive_actions_plot = sns.barplot(data = barca_defend, x = "Tkl+Int", y = "Player", hue = "Pos")
defensive_actions_plot.set(ylabel = None)
plt.xlabel("Total Tackles and Interceptions")
plt.title("Barcelona Players Defensive Actions in 2022-23", pad = 15, color = "#333333", weight = "bold")
plt.show()

# other metrics:
# Blocks (can combine tackles + interceptions + blocks)
# Mid 3rd (tackles won in the mid 3rd)
# Tkl% (dribblers tackled percentage / one on one success rate)
```

```python
barca_defend = barca_defend.sort_values(by = ["Mid 3rd"], ascending = False)

#fig = plt.figure(facecolor='#efeceb')
sns.set(style="darkgrid")
mid_3rd_plot = sns.barplot(data = barca_defend, x = "Mid 3rd", y = "Player", hue = "Pos", palette = 'GnBu_r')
mid_3rd_plot.set(ylabel = None)
plt.xlabel("Tackles in Middle 1/3")
plt.title("Barcelona Players Tackles in Middle 1/3 in 2022-23", pad = 15, color = "#333333", weight = "bold")
plt.show()
```

```python
barca_defend = barca_defend.sort_values(by = ["DribTkl"], ascending = False)

fig = plt.figure(facecolor='#efeceb')
palette = [ '#08589e','#2b8cbe', '#7bccc4' ]
sns.set(style="darkgrid")
drib_tkl_plot = sns.barplot(data = barca_defend, x = "DribTkl", y = "Player", hue = "Pos", palette = palette)
drib_tkl_plot.set(ylabel = None)
plt.xlabel("Number of Dribblers Tackled")
plt.title("Dribblers Tackled by Barcelona Players in 2022-23", pad = 15, color = "#333333", weight = "bold")
plt.show()
```
